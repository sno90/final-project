```python
# import dependencies
import pandas as pd
import numpy as np
```

```python
# import files 
unique_chems = pd.read_csv('data/Unique_Chem.csv', usecols=['ChemicalName'])
ulta_products_df = pd.read_csv('data/ulta_makeup.csv',usecols= ['brand', 'description', 'ingredients list'])
CA_products_df = pd.read_csv('data/CA_Products.csv', usecols=['BrandName', 'ProductName', 'ChemicalName', 'Discontinued'])
```

```python
ulta_products_df['Discontinued'] = 0

ulta_products_df.head()
```

```python
CA_products_df = CA_products_df[['BrandName', 'ProductName', 'ChemicalName', 'Discontinued']]
CA_products_df.rename(columns = {'BrandName':'brand', 'ProductName':'description', 
                              'ChemicalName':'ingredients list', 'Discontinued':'Discontinued'}, inplace = True) 

CA_products_df.head()
```

```python
# combined_products_df = ulta_products_df.append(CA_products_df)
combined_products_df = pd.concat([ulta_products_df, CA_products_df], axis=0)
```

```python
combined_products_df.head()
```

```python
# create columns in the ulta df for each toxic chem
for chem in unique_chems['ChemicalName']:
    combined_products_df[chem] = 0
```

```python
combined_products_df.reset_index(inplace=True)
```

```python
# loop through each products ingredients list to see if the toxic chemicals are included 
for chem in unique_chems['ChemicalName']: 
    for i in range(0, len(combined_products_df)-1):
        if chem in combined_products_df['ingredients list'][i]:
            combined_products_df[chem][i] = 1
        else:
            'None'
            
```
