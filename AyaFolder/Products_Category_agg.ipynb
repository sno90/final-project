```python
import pandas as pd
```

```python
csv_file = "cscpopendata.csv"
prodChemicals_df = pd.read_csv(csv_file)
prodChemicals_df.head()
```

```python
prodChemicals_df = prodChemicals_df.dropna(subset=['DiscontinuedDate'])
prodChemicals_df.head()
```

```python
prodChemicals_df = prodChemicals_df.groupby('PrimaryCategory').agg({'DiscontinuedDate':','.join,'DiscontinuedDate':"count", 'ProductName':','.join})
prodChemicals_df
```

```python
prodChemicals_df = prodChemicals_df.sort_values(by='DiscontinuedDate', ascending=False)
prodChemicals_df
```

```python
prodChemicals_df.to_csv("Products_Category_Ascending.csv")
```
