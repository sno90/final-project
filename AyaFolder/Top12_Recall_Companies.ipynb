```python
import pandas as pd
```

```python
csv_file = "cscpopendata.csv"
cosmeticData_df = pd.read_csv(csv_file)
cosmeticData_df.head()
```

```python
cosmeticData_df = cosmeticData_df.dropna(subset=['DiscontinuedDate'])
cosmeticData_df.head()
```

```python
# cosmeticData_df['Counts'] = cosmeticData_df.groupby(['CompanyName'])['DiscontinuedDate'].transform('count')
# cosmeticData_df.head()

cosmeticData_df = cosmeticData_df.groupby('CompanyName').agg({'DiscontinuedDate':','.join, 'ChemicalName': ', '.join, 'DiscontinuedDate':"count"}).reset_index()
cosmeticData_df.head()
```

```python
cosmeticData_df = cosmeticData_df.sort_values(by='DiscontinuedDate', ascending=False)
cosmeticData_df
```

```python
cosmeticData_df.to_csv("All_Companies_Ascending_DD.csv")
```
