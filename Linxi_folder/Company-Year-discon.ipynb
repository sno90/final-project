```python
import pandas as pd
```

```python
csv_file = "../AyaFolder/cscpopendata.csv"
raw_brand = pd.read_csv(csv_file)
```

```python
raw_brand = raw_brand.loc[raw_brand['CompanyName'].isin( ["The Procter & Gamble Company","Victoria's Secret Beauty","Bath & Body Works","GUERLAIN","Too Faced Cosmetics","Smashbox Cosmetics","Shiseido Co., Ltd.","New Avon LLC"])]
                                                         
                          
```

```python
raw_brand.dtypes
```

```python
raw_brand['DiscontinuedDate'] = pd.to_datetime(raw_brand['DiscontinuedDate'])
```

```python
raw_brand['Year'] = raw_brand['DiscontinuedDate']
raw_brand
```

```python
raw_brand['Year'] = raw_brand['Year'].dt.strftime('%Y')
```

```python
df = raw_brand[['ProductName','CompanyName','DiscontinuedDate','Year']]
```

```python
df
```

```python
df1 = df.dropna(subset=["DiscontinuedDate"])
df1
```

```python
df2 = df1.groupby(['CompanyName','Year']).count()
```
